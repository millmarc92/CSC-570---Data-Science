```python
import pandas as pd
import networkx as nx
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
co_occurence = pd.read_csv("co_occurence.csv", index_col='Unnamed: 0')
```

```python
co_occurence
```

```python
G = nx.from_numpy_matrix(co_occurence.values)
G = nx.relabel_nodes(G, dict(enumerate(co_occurence.columns)))
```

```python
plt.figure(figsize=(40,40))
pos=nx.spring_layout(G, iterations=500, scale=5, k=.3)
nx.draw_networkx_labels(G,pos,fontsize=8)
font = {'fontname'   : 'Helvetica',
            'color'      : 'k',
            'fontweight' : 'bold',
            'fontsize'   : 14}
plt.title("Boston Citizens 1776", font)


nx.draw_networkx_nodes(G,pos,node_color='b',alpha=0.4)
nx.draw_networkx_edges(G,pos,alpha=0.4,node_size=0,width=1,edge_color='k')
plt.axis('off')
plt.show()
```

```python
nx.write_gexf(G,'medline.gexf')
```
