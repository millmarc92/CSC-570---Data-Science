```python
from PIL import Image
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

### Some things about images...

I'm going to use two different python packages for Image manipulation.  Python Image Library (PIL) has lots of great features for working with images.   Matplotlib's Pyplot has some tools that help as well.   

I'll also be using numpy to show you what images look like in their raw form.  


```python
#PIL.Image.open can read an image from disk
im=Image.open("luna.jpg")
```

```python
#and pyplot's imshow works great to display an image in a notebook.

plt.imshow(im)
```

Meet my cat, Luna.

It's very common to resize images to some consistent size in computer vision work.  Here I'll resize Luna to 256*256 pixels.

This makes Luna look quite square and fat, like her sister Eve.

```python
im = im.resize((256,256))
plt.imshow(im)
```

### Pictures are just another matrix...

Although their shape may seem odd.  Color pictures are a matrix with a shape of Pixel Width * Pixel Height * 3

You can think of this as a 3D matrix, or three matricies of Pixel Width * Pixel Height stacked on top of each other.   The word Tensor is often used to desribe these matricies very generally.

```python
cat_array = np.array(im)
```

```python
cat_array.shape
```

The first two dimensions are picture width and height.   

```python
#if we select only 
cat_array[:,:,0]
```

```python
cat_array[:,:,0].shape
```

The third dimension is used for color channels.   We can easily look at only one channel

```python
plt.imshow(cat_array[:,:,0])
```

```python
plt.imshow(cat_array)
```
